```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('bisecting').getOrCreate()
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/iris.csv"

iris = spark.read.csv(abs_path, header=True, inferSchema=True)
iris.show(5)
```

```python
from pyspark.ml.clustering import BisectingKMeans
```

```python
from pyspark.ml.feature import VectorAssembler

asb = VectorAssembler(inputCols=iris.columns[:-1], outputCol='features')
iris_ass = asb.transform(iris)

iris_ass.show(5)
```

```python
from pyspark.ml.clustering import BisectingKMeans
from pyspark.ml.feature import VectorAssembler

iris = spark.read.csv(
    "/Volumes/workspace/ml_with_spark/raw_data/iris.csv",
    header=True,
    inferSchema=True
)

asb = VectorAssembler(
    inputCols=["sepallength", "sepalwidth", "petallength", "petalwidth"],
    outputCol="features"
)

iris_ass = asb.transform(iris)

iris_ass.write.mode("overwrite").saveAsTable("workspace.ml_with_spark.iris_features")

iris_ass = spark.table("workspace.ml_with_spark.iris_features")

bkm = BisectingKMeans(
    featuresCol="features",
    predictionCol="cluster",
    k=3
)

modelo_bkm = bkm.fit(iris_ass)

clusters = modelo_bkm.transform(iris_ass)

clusters.show(5)
```
