# Univariate Feature Selector

Nem sempre as features ou targets são da mesma natureza; temos features/targets categóricos e contínuos.

O Spark pode usar uma técnica diferente dependendo da combinação feature x target:
- Feat categórica e Target contínuo: qui-quadrado
- Feat contínuo e Target categórico: ANOVA
- Feat contínuo e Target contínuo: Estatística F

Parâmetros:
- featureType: tipo de feature -> categórico ou contínuo
- labelType: tipo de target
- selectionThreshold: limite de características que serão selecionadas

```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('univariate_selector').getOrCreate()
```

```python
from pyspark.ml.feature import ChiSqSelector, RFormula, UnivariateFeatureSelector
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Carros.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

```python
rformula = RFormula(formula='HP ~ .', featuresCol='independente', labelCol='dependente')
carros_rf = rformula.fit(carros).transform(carros)
carros_rf.select('independente', 'dependente').show(5, truncate=False)
```

```python
selector = UnivariateFeatureSelector(
    featuresCol='independente',
    outputCol='selecionados',
    labelCol='dependente'
)
selector.setFeatureType('continuous').setLabelType('continuous').setSelectionThreshold(5)
carros_uni = selector.fit(carros_rf).transform(carros_rf)
carros_uni.select('selecionados').show(5, truncate=False)
```
