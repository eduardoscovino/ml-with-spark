# Quantile Discretizer

Técnica de transformação de dados contínuos em discretos.

Parâmetro:
- buckets: quantos valores discretos serão usados (não necessariamente todos serão usados)

```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('quantile_discretizer').getOrCreate()
```

```python
from pyspark.ml.feature import QuantileDiscretizer, VectorAssembler
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Churn.csv"

churn = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
churn.show(5)
```

```python
discrete = QuantileDiscretizer(numBuckets=4, inputCol='Tenure', outputCol='discrete_Tenure')

discrete_churn = discrete.fit(churn).transform(churn)

discrete_churn.select('Tenure', 'discrete_Tenure').show()
```
