# Geração e extração de features: PCA

Uma alta dimensionalidade aumenta o risco de overfitting e torna o treinamento mais lento.
-> O PCA (redução de dimensionalidade) cria features sintéticas, sem compreensão funcional, mas reduz o número de features
Não permite avaliar importância de features e não mais representam o negócio analisado.

```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('pca').getOrCreate()
```

```python
from pyspark.ml.feature import PCA
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Carros.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

Primeiro precisa colocar as features numa única coluna com um vetor contendo todas elas

```python
from pyspark.ml.feature import VectorAssembler

vectas = VectorAssembler(inputCols= ["Consumo", "Cilindros", "Cilindradas", "RelEixoTraseiro", "Peso", "Tempo", "TipoMotor", "Transmissao", "Marchas", "Carburadors"], outputCol='features')

carros = vectas.transform(carros)
carros.select('features').show(truncate=False)
```

```python
pca = PCA(k=3, inputCol='features', outputCol='pcaFeatures')
model = pca.fit(carros)

carros_pca = model.transform(carros)

display(carros_pca)
```

```python
carros_pca.select('pcaFeatures').show(truncate=False)
```
