```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('ohe').getOrCreate()
```

```python
from pyspark.ml.feature import IndexToString, StringIndexer, OneHotEncoder
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Churn.csv"

churn = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
churn.show(5)
```

```python
string_index = StringIndexer(inputCol='Geography', outputCol='Indexer_c1')
index_ohe = string_index.fit(churn).transform(churn)

string_index = StringIndexer(inputCol='Gender', outputCol='Indexer_c2')
index_ohe = string_index.fit(index_ohe).transform(index_ohe)
```

```python
display(index_ohe)
```

```python
ohe = OneHotEncoder(inputCols=['Indexer_c1', 'Indexer_c2'], outputCols=['Geography_ohe', 'Gender_ohe'])
ohe_model = ohe.fit(index_ohe)
churn_ohe = ohe_model.transform(index_ohe)
```

indices e values
-> o OHE no spark devolve um vetor esparso.
-> ele dropa a última label: em Geography é Spain. Por isso que nas linhas de Spain o vetor é vazio de size 2, pois sobraram duas labels (France e Germany)

France:
value 1 e indice 0 -> tem o valor 1 no índice 0 do vetor

Germany:
value 1 e indice 1 -> tem o valor 1 no índice 1 do vetor

Spain:
value [] e indice [] -> vetor de size 2 esparso, não tem nada preenchido

Por default, o OHE dropa o valor do último label, mas pode mudar para dropar o primeiro.
