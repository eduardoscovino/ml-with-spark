```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('index_to_string').getOrCreate()
```

```python
from pyspark.ml.feature import IndexToString, StringIndexer
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Churn.csv"

churn = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
churn.show(5)
```

```python
string_indexer = StringIndexer(inputCol='Geography', outputCol='Geography_index')
string_indexer_model = string_indexer.fit(churn)
churn_indexed = string_indexer_model.transform(churn)

churn_indexed.select('Geography', 'Geography_index').show(5)
```

```python
string_indexer_model.labels
```

```python
retorno = IndexToString(inputCol='Geography_index', outputCol='Geography_original')
churn_return = retorno.transform(churn_indexed)

display(churn_return)
```
