```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('polynomial_expansion').getOrCreate()
```

```python
from pyspark.ml.feature import IndexToString, StringIndexer, OneHotEncoder, Imputer, PolynomialExpansion
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Carros.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler

vec_assembler = VectorAssembler(inputCols=['Consumo', 'Cilindros', 'Cilindradas'], outputCol='vector')
carros_vector = vec_assembler.transform(carros)
carros_vector.show(5)
```

```python
polynomial = PolynomialExpansion(inputCol='vector', outputCol='poly_features', degree=2)
carros_poly = polynomial.transform(carros_vector)
carros_poly.select('vector', 'poly_features').show(truncate=False)
```
