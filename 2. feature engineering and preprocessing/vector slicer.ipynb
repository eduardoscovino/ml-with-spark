# Vector Slicer

- Recebe uma coluna com um vetor de atributos
- Cria uma nova coluna com os atributos especificados pelo índice

```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('vector_slicer').getOrCreate()
```

```python
from pyspark.ml.feature import QuantileDiscretizer, VectorAssembler, RFormula, VectorSlicer
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Carros.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

```python
rformula = RFormula(formula='HP ~ .', featuresCol='independente', labelCol='dependente')
carros_rf = rformula.fit(carros).transform(carros)
carros_rf.show(5)
```

```python
slicer = VectorSlicer(inputCol='independente', outputCol='slicer_indep', indices=[1, 2, 6])
carros_slicer = slicer.transform(carros_rf)
carros_slicer.select('dependente', 'independente', 'slicer_indep').show(truncate=False)
```
