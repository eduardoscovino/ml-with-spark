```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('binarizer').getOrCreate()
```

```python
from pyspark.ml.feature import Binarizer
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/iris.csv"

iris = spark.read.csv(abs_path, header=True, inferSchema=True, sep=',')
iris.show(5)
```

```python
binarizer = Binarizer(threshold=5, inputCol='sepallength', outputCol='sepallength_bin')

irisbin = binarizer.transform(iris)

display(irisbin)
```
