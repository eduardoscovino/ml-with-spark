```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('maxabs_scaler').getOrCreate()
```

```python
from pyspark.ml.feature import IndexToString, StringIndexer, OneHotEncoder, Imputer, MaxAbsScaler
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/Carros.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

```python
from pyspark.ml.feature import VectorAssembler

vec_assembler = VectorAssembler(inputCols=['Consumo', 'Cilindros', 'Cilindradas'], outputCol='matriz')
carros_vector = vec_assembler.transform(carros)
carros_vector.show(5)
```

```python
mabs_scaler = MaxAbsScaler(inputCol='matriz', outputCol='mabs_matriz')
mabs_scaler_model = mabs_scaler.fit(carros_vector)
carros_std = mabs_scaler_model.transform(carros_vector)
carros_std.select('matriz', 'mabs_matriz').show(truncate=False)
```
