```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('imputer').getOrCreate()
```

```python
from pyspark.ml.feature import IndexToString, StringIndexer, OneHotEncoder, Imputer
```

```python
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/CarrosNAN.csv"

carros = spark.read.csv(abs_path, header=True, inferSchema=True, sep=';')
carros.show(5)
```

```python
imput = Imputer(inputCols=['Cilindradas', 'Peso'], outputCols=['novaCilindradas', 'novaPeso'])
imput_model = imput.fit(carros)
carros_imput = imput_model.transform(carros)
carros_imput.select('Cilindradas', 'novaCilindradas', 'Peso', 'novaPeso').show(5)
```

```python
imput = Imputer(inputCols=['Cilindros'], outputCols=['novaCilindros'])
imput.setStrategy('median').setMissingValue(0).fit(carros).transform(carros).select('Cilindros', 'novaCilindros').show()
```
