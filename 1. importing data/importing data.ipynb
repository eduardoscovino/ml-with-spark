# Parte 1

```python
import findspark, pyspark
from pyspark.sql import SparkSession
findspark.init()
spark = SparkSession.builder.appName('importacao').getOrCreate()
```

```python
arqschema = "id INT, nome STRING, status STRING, cidade STRING, vendas INT, data DATE"
abs_path = "/Volumes/workspace/ml_with_spark/raw_data/despachantes.csv"

despachantes = (
    spark.read
    .option("header", True)
    .schema(arqschema)
    .csv(abs_path)
)

despachantes.show()
```

```python
# metodo mais generico para puxar qualquer tipo de formato
desp_autoschema = spark.read.load(abs_path, format='csv', header=False, inferSchema=True, sep=',')
desp_autoschema.show()
```

```python
desp_autoschema.schema
```

```python
par_path = "/Volumes/workspace/ml_with_spark/raw_data/despachantes.parquet"
par = spark.read.format('parquet').load(par_path)
par.show()
```

```python
js_path = "/Volumes/workspace/ml_with_spark/raw_data/despachantes.json"
js = spark.read.format('json').load(js_path)
js.show()
```

```python
orc_path = "/Volumes/workspace/ml_with_spark/raw_data/despachantes.orc"
orc = spark.read.format('orc').load(orc_path)
orc.show()
```
